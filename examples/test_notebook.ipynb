```python
from NoisyCircuits import QuantumCircuit as QC
from NoisyCircuits.utils.GetNoiseModel import GetNoiseModel
import pickle
import os
import json
```

```python
token = json.load(open(os.path.join(os.path.expanduser("~"), "ibm_api.json"), "r"))["apikey"] # Replace with your IBM Quantum token
backend_name = "ibm_fez"
num_qubits = 9
num_cores = 1
num_trajectories = 20
threshold = 1e-4
jsonize = True
verbose = True
# Options: "heron", "eagle". Note that "eagle" is now deprecated and only available in simulation mode (i.e., no noise model from hardware).
qpu_type = "heron" 
```

```python
noise_model = pickle.load(open("../noise_models/Noise_Model_Heron_QPU.pkl", "rb"))
```

```python
nqc = QC(num_qubits=num_qubits, 
         noise_model=noise_model, 
         num_cores=num_cores,
         backend_qpu_type="heron", 
         num_trajectories=num_trajectories, 
         threshold=threshold, 
         jsonize=jsonize,
         verbose=verbose)
```

```python
for q_num in nqc.single_qubit_instructions:
    for op_name in ["x", "sx"]:
        zeros_ratio_list = []
        print(f"Analyzing {op_name} gate on qubit {q_num}:-")
        ops = nqc.single_qubit_instructions[q_num][op_name]["kraus_operators"]
        for op in ops:
            sys_size = 2 ** num_qubits
            zero_count = 0
            for i in range(sys_size):
                for j in range(sys_size):
                    if op[i, j] == 0:
                        zero_count += 1
            zeros_ratio_list.append(zero_count / (sys_size * sys_size))
        print(f"Average zeros ratio in Kraus operators:", sum(zeros_ratio_list) / len(zeros_ratio_list))
```

```python
for key_pair in nqc.two_qubit_instructions["cz"].keys():
    zeros_ratio_list = []
    ops = nqc.two_qubit_instructions["cz"][key_pair]["operators"]
    for op in ops:
        sys_size = 2 ** num_qubits
        zero_count = 0
        for i in range(sys_size):
            for j in range(sys_size):
                if op[i, j] == 0:
                    zero_count += 1
        zeros_ratio_list.append(zero_count / (sys_size * sys_size))
    print(f"Average zeros ratio in Kraus operators of CZ gate on qubits {key_pair}:", sum(zeros_ratio_list) / len(zeros_ratio_list))
```

```python
from scipy.sparse import csr_matrix, csc_matrix, issparse
import numpy as np
```

```python
for q_num in nqc.single_qubit_instructions:
    for op_name in ["x", "sx"]:
        zeros_ratio_list = []
        print(f"Analyzing {op_name} gate on qubit {q_num}:-")
        ops = nqc.single_qubit_instructions[q_num][op_name]["kraus_operators"]
        sparse_op_list = []
        for op in ops:
            if issparse(op):
                sparse_op_list.append(True)
            else:
                sparse_op_list.append(False)
        print(f"Sparse representation in Kraus operators:", np.all(sparse_op_list))
```
